````python
# seed кі м •
import random
import torch.backends.cudnn as cudnn

def random_seed(seed_num):
    torch.manual_seed(seed_num)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed_num)
        torch.cuda.manual_seed_all(seed_num)
        cudnn.benchmark = False
        cudnn.deterministic = True
    np.random.seed(seed_num)
    random.seed(seed_num)

random_seed(42)
````

#### рҹ“қ м„ӨлӘ…: н…ҚмҠӨнҠё лҚ°мқҙн„° м „мІҳлҰ¬н•ҳкё°
* н•ҙлӢ№ лҚ°мқҙн„°м…ӢмқҖ нҒ¬лЎӨл§Ғ(мқён„°л„·м—җ мһҲлҠ” м •ліҙлҘј мҲҳм§‘н•ҳлҠ” кё°лІ•)мқ„ нҶөн•ҙ кө¬м¶•лҗҳм—Ҳкё° л•Ңл¬ём—җ no-break spaceк°Җ мў…мў… л°ңмғқн•©лӢҲлӢӨ. мқҙлҹ¬н•ң no-break spaceлҘј м ңкұ°н•ҳлҠ” м „мІҳлҰ¬лҘј м§„н–үн•©лӢҲлӢӨ.
  * No-Break SpaceлһҖ? мӣ№ нҺҳмқҙм§ҖлӮҳ л¬ём„ң л“ұм—җм„ң лӢЁм–ҙлӮҳ л¬ёмһҘ мӮ¬мқҙмқҳ кіөл°ұмқҙ мһҲлҠ” кІҪмҡ°, н•ҙлӢ№ кіөл°ұмқҙ мӨ„ л°”кҝҲмңјлЎң мқён•ҙ л¶„лҰ¬лҗҳм§Җ м•Ҡкі  н•ң лӢЁм–ҙлӮҳ л¬ёмһҘмңјлЎң мқёмӢқлҗҳлҸ„лЎқ н•ҳлҠ”лҚ° мӮ¬мҡ©лҗҳлҠ” кіөл°ұ
  * мҳҲмӢң (no-break-space мӮ¬мҡ© X)
    ```
    Hello
    World~
    ```
    
    (no-break-space мӮ¬мҡ©)
    ```
    Hello,вҺөworld!
    ```
* no-break spaceлҘј м ңкұ°н•ҳкё° мң„н•ҙм„  unicode нҳ•мӢқмңјлЎң м ңкұ°лҘј н•ҙм•јн•©лӢҲлӢӨ.
  * unicodeлһҖ? м „ м„ёкі„мқҳ лӘЁл“  л¬ёмһҗмҷҖ кё°нҳёлҘј мқјкҙҖм„ұ мһҲкІҢ н‘ңнҳ„н•ҳкё° мң„н•ң н‘ңмӨҖ л¬ёмһҗ мқёмҪ”л”© мІҙкі„

* `re` лқјмқҙлёҢлҹ¬лҰ¬лҘј мқҙмҡ©н•ҳм—¬ н…ҚмҠӨнҠё лҚ°мқҙн„°лҘј мүҪкІҢ мІҳлҰ¬н•  мҲҳ мһҲмҠөлӢҲлӢӨ.

рҹ“ҡ м°ёкі н• л§Ңн•ң мһҗлЈҢ:
* <a href='https://www.compart.com/en/unicode'>unicode кІҖмғү мӮ¬мқҙнҠё</a>
* [re лқјмқҙлёҢлҹ¬лҰ¬лҘј мқҙмҡ©н•ң н…ҚмҠӨнҠё лҚ°мқҙн„° мӮ¬мҡ©лІ•](https://velog.io/@hoegon02/%EC%9E%90%EC%97%B0%EC%96%B4%EC%B2%98%EB%A6%AC-12-%ED%85%8D%EC%8A%A4%ED%8A%B8-%EC%A0%84%EC%B2%98%EB%A6%AC-%EC%A0%95%EA%B7%9C-%ED%91%9C%ED%98%84%EC%8B%9D-3qmtwryf)

````python
def cleaning_text(text):
    cleaned_text = re.sub( r"[^a-zA-Z0-9.,@#!\s']+", "", text) # нҠ№мҲҳл¬ёмһҗ лҘј лӘЁл‘җ м§Җмҡ°лҠ” мһ‘м—…мқ„ мҲҳн–үн•©лӢҲлӢӨ.
    cleaned_text = cleaned_text.replace(u'\xa0',u' ') # No-break spaceлҘј unicode л№Ҳм№ёмңјлЎң ліҖнҷҳ
    cleaned_text = cleaned_text.replace('\u200a',' ') # unicode л№Ҳм№ёмқ„ л№Ҳм№ёмңјлЎң ліҖнҷҳ
    return cleaned_text

cleaned_data = list(map(cleaning_text, data)) # лӘЁл“  нҠ№мҲҳл¬ёмһҗмҷҖ кіөл°ұмқ„ м§ҖмӣҖ
print('Before preprocessing')
print(data[:5])
print('After preprocessing')
print(cleaned_data[:5])
````

#### рҹ“қ м„ӨлӘ…: Tokenizer

````python
# лӢЁм–ҙ мӮ¬м „мқ„ torchtext м—Ҷмқҙ м§Ғм ‘ мғқм„ұн•ң нӣ„, pad нҶ нҒ°мқ„ м¶”к°Җн•©лӢҲлӢӨ.

from collections import Counter

# лӘЁл“  нҶ нҒ° м¶”м¶ң
all_tokens = []
for sentence in cleaned_data:
    all_tokens.extend(tokenizer(sentence))

# нҶ нҒ° л№ҲлҸ„ кі„мӮ°
token_freq = Counter(all_tokens)

# кі мң  нҶ нҒ° лӘ©лЎқ м¶”м¶ң л°Ҹ <pad> нҶ нҒ°мқ„ мІ« лІҲм§ёлЎң мӮҪмһ…
unique_tokens = ['<pad>'] + sorted(token_freq.keys())

# нҶ нҒ°мқ„ мқёлҚұмҠӨлЎң л§Өн•‘ (string to id)
token2id = {token: idx for idx, token in enumerate(unique_tokens)}

# idлҘј нҶ нҒ°мңјлЎң л§Өн•‘ (id to string)
id2token = {idx: token for token, idx in token2id.items()}

# vocabulary к°қмІҙмІҳлҹј мӮ¬мҡ©н•  мҲҳ мһҲлҠ” к°„лӢЁн•ң нҒҙлһҳмҠӨ
class Vocab:
    def __init__(self, token2id, id2token):
        self._token2id = token2id
        self._id2token = id2token

    def lookup_indices(self, tokens):
        return [self._token2id.get(token, self._token2id['<pad>']) for token in tokens]

    def get_itos(self):
        return [self._id2token[i] for i in range(len(self._id2token))]

    def get_stoi(self):
        return self._token2id

vocab = Vocab(token2id, id2token)
````

#### рҹ“қ м„ӨлӘ…: Custom Dataset м •мқҳн•ҳкё°
* 1-1м—җм„ң м§„н–үн•ң м „мІҳлҰ¬ кіјм •мқ„ лӘЁл‘җ н•ЁмҲҳнҷ” мӢңмјңм„ң н•ҳлӮҳмқҳ classлЎң кө¬м¶•н•©лӢҲлӢӨ.

* лҚ°мқҙн„°лҘј ліҖнҷҳн•ҳлҠ” кіјм •мқҖ лҗҳлҸ„лЎқмқҙл©ҙ getitem мқҙ м•„лӢҢ init л¶Җл¶„м—җ н•ҳм—¬, м „мІҳлҰ¬н•ҳлҠ” мӢңк°„мқ„ мӨ„мқҙлҸ„лЎқ н•©лӢҲлӢӨ.
  * init л¶Җл¶„м—җ н•ң лІҲм—җ н•ҳкІҢ лҗҳл©ҙ datasetмқ„ м •мқҳн•  л•Ңл§Ң ліҖнҷҳ мӢңк°„мқҙ мҶҢмҡ”лҗҳкі , к·ё мқҙнӣ„лЎңлҠ” лҚ°мқҙн„° м „мІҳлҰ¬ мӢңк°„мқҙ мҶҢмҡ”лҗҳм§Җ м•ҠмҠөлӢҲлӢӨ.

рҹ“ҡ м°ёкі н• л§Ңн•ң мһҗлЈҢ:
* [Custom Dataset кө¬м¶• - Pytorch кіөмӢқ нҠңнҶ лҰ¬м–ј](https://tutorials.pytorch.kr/beginner/basics/data_tutorial.html)

````python
class CustomDataset(Dataset):
    def __init__(self, data, vocab, tokenizer, max_len):
        self.data = data
        self.vocab = vocab
        self.max_len = max_len
        self.tokenizer = tokenizer
        seq = self.make_sequence(self.data, self.vocab, self.tokenizer) # next word predictionмқ„ н•ҳкё° мң„н•ң нҳ•нғңлЎң ліҖнҷҳ
        self.seq = self.pre_zeropadding(seq, self.max_len) # zero paddingмңјлЎң мұ„мӣҢмӨҢ
        self.X = torch.tensor(self.seq[:,:-1])
        self.label = torch.tensor(self.seq[:,-1])

    def make_sequence(self, data, vocab, tokenizer):
        seq = []
        for i in data:
            token_id = vocab.lookup_indices(tokenizer(i))
            for j in range(1, len(token_id)):
                sequence = token_id[:j+1]
                seq.append(sequence)
        return seq

    def pre_zeropadding(self, seq, max_len): # max_len кёёмқҙм—җ л§һм¶°м„ң 0 мңјлЎң padding мІҳлҰ¬ (м•һл¶Җл¶„м—җ padding мІҳлҰ¬)
        return np.array([i[:max_len] if len(i) >= max_len else [0] * (max_len - len(i)) + i for i in seq])

    def __len__(self): # datasetмқҳ м „мІҙ кёёмқҙ л°ҳнҷҳ
        return len(self.X)

    def __getitem__(self, idx): # dataset м ‘к·ј
        X = self.X[idx]
        label = self.label[idx]
        return X, label
````

````python
import re
from collections import Counter

def cleaning_text(text):
    cleaned_text = re.sub( r"[^a-zA-Z0-9.,@#!\s']+", "", text) # нҠ№мҲҳл¬ёмһҗ лҘј лӘЁл‘җ м§Җмҡ°лҠ” мһ‘м—…мқ„ мҲҳн–үн•©лӢҲлӢӨ.
    cleaned_text = cleaned_text.replace(u'\xa0',u' ') # No-break spaceлҘј unicode л№Ҳм№ёмңјлЎң ліҖнҷҳ
    cleaned_text = cleaned_text.replace('\u200a',' ') # unicode л№Ҳм№ёмқ„ л№Ҳм№ёмңјлЎң ліҖнҷҳ
    return cleaned_text

data = list(map(cleaning_text, data))

def simple_tokenizer(text):
    # basic_englishкіј мң мӮ¬н•ҳкІҢ мҶҢл¬ёмһҗ ліҖнҷҳ нӣ„, кіөл°ұ л°Ҹ нҠ№мҲҳл¬ёмһҗ кё°мӨҖ нҶ нҒ°нҷ”
    text = text.lower()
    # лӢЁм–ҙ, мҲ«мһҗ, . , , @ # ! ' мқ„ кё°мӨҖмңјлЎңл§Ң лӮҳлҲ„кІҢлҒ” лӢЁмҲң split
    return text.split()

tokenizer = simple_tokenizer

# Vocabulary м§Ғм ‘ мғқм„ұ
counter = Counter()
for line in data:
    counter.update(tokenizer(line))
# pad нҶ нҒ° мҡ°м„  м¶”к°Җ
vocab_list = ['<pad>'] + [word for word, _ in counter.most_common()]
# нҶ нҒ° -> мқёлҚұмҠӨ, мқёлҚұмҠӨ -> нҶ нҒ° мӮ¬м „
class SimpleVocab:
    def __init__(self, vocab_list):
        self.itos = list(vocab_list)
        self.stoi = {tok: idx for idx, tok in enumerate(self.itos)}
    def lookup_indices(self, tokens):
        return [self.stoi.get(tok, 0) for tok in tokens]  # unknownмқҖ <pad>(0)лЎң мІҳлҰ¬

vocab = SimpleVocab(vocab_list)
max_len = 20
````

````python
class NextWordPredictionModel(nn.Module):
    def __init__(self, vocab_size, embedding_dims, hidden_dims, num_classes, dropout_ratio, set_super):
        if set_super:
            super().__init__()

        self.embedding = nn.Embedding(vocab_size, embedding_dims, padding_idx = 0) # padding index м„Өм • => gradient кі„мӮ°м—җм„ң м ңмҷё
        self.hidden_dims = hidden_dims
        self.layers = nn.ModuleList()
        self.num_classes = num_classes
        for i in range(len(self.hidden_dims) - 1):
            self.layers.append(nn.Linear(self.hidden_dims[i], self.hidden_dims[i+1]))

            self.layers.append(nn.BatchNorm1d(self.hidden_dims[i+1]))

            self.layers.append(nn.ReLU())

            self.layers.append(nn.Dropout(dropout_ratio))

        self.classifier = nn.Linear(self.hidden_dims[-1], self.num_classes)
        self.softmax = nn.LogSoftmax(dim = 1)

    def forward(self, x):
        '''
        INPUT:
            x: [batch_size, sequence_len] # padding м ңмҷё
        OUTPUT:
            output : [batch_size, vocab_size]
        '''
        x = self.embedding(x) # [batch_size, sequence_len, embedding_dim]
        x = torch.sum(x, dim=1) # [batch_size, embedding_dim] к°Ғ л¬ёмһҘм—җ лҢҖн•ҙ мһ„лІ л”©лҗң лӢЁм–ҙл“Өмқ„ н•©міҗм„ң, н•ҙлӢ№ л¬ёмһҘм—җ лҢҖн•ң мһ„лІ л”© лІЎн„°лЎң л§Ңл“Өм–ҙмӨҚлӢҲлӢӨ.
        for layer in self.layers:
            x = layer(x)

        output = self.classifier(x) # [batch_size, num_classes]
        output = self.softmax(output) # [batch_size, num_classes]
        return output

    def count_parameters(self):
        return sum(p.numel() for p in self.parameters() if p.requires_grad)
````

#### рҹ“қ м„ӨлӘ…: Next word prediction н•ҷмҠөн•ҳкё°
* DNN лӘЁлҚёмқ„ н•ҷмҠөн•ҳкё° мң„н•ҙ лӘЁлҚёмқҳ нҢҢлқјлҜён„°лҘј м •н•ҙмӨҚлӢҲлӢӨ.
* embedding layerмҷҖ fully connected layerмқҳ м—°мӮ°мқҙ к°ҖлҠҘн•ҳкІҢ н•ҳкё° мң„н•ҙ hidden dimension лҰ¬мҠӨнҠё кө¬м„ұ мӢң, embedding dimensionмқ„ мІ«лІҲм§ё к°’мңјлЎң м„Өм •н•©лӢҲлӢӨ.
* мҳҲмёЎн•ҳл ӨлҠ” labelмқҳ к°ңмҲҳлҠ” лӢЁм–ҙ мӮ¬м „м—җ мһҲлҠ” лӢЁм–ҙмқҳ к°ңмҲҳмҷҖ лҸҷмқјн•©лӢҲлӢӨ.

````python
# training мҪ”л“ң, evaluation мҪ”л“ң, training loop мҪ”л“ң
def training(model, dataloader, train_dataset, criterion, optimizer, device, epoch, num_epochs):
    model.train()  # лӘЁлҚёмқ„ н•ҷмҠө лӘЁл“ңлЎң м„Өм •
    train_loss = 0.0
    train_accuracy = 0

    tbar = tqdm(dataloader)
    for texts, labels in tbar:
        texts = texts.to(device)
        labels = labels.to(device)

        # мҲңм „нҢҢ
        outputs = model(texts)

        loss = criterion(outputs, labels)

        # м—ӯм „нҢҢ л°Ҹ к°ҖмӨ‘м№ҳ м—…лҚ°мқҙнҠё
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # мҶҗмӢӨкіј м •нҷ•лҸ„ кі„мӮ°
        train_loss += loss.item()
        # torch.maxм—җм„ң dim мқёмһҗм—җ к°’мқ„ м¶”к°Җн•  кІҪмҡ°, н•ҙлӢ№ dimensionм—җм„ң мөңлҢ“к°’кіј мөңлҢ“к°’м—җ н•ҙлӢ№н•ҳлҠ” мқёлҚұмҠӨлҘј л°ҳнҷҳ
        _, predicted = torch.max(outputs, dim=1)


        train_accuracy += (predicted == labels).sum().item()

        # tqdmмқҳ м§„н–үл°”м—җ н‘ңмӢңлҗ  м„ӨлӘ… н…ҚмҠӨнҠёлҘј м„Өм •
        tbar.set_description(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {loss.item():.4f}")

    # м—җнҸӯлі„ н•ҷмҠө кІ°кіј м¶ңл Ҙ
    train_loss = train_loss / len(dataloader)
    train_accuracy = train_accuracy / len(train_dataset)

    return model, train_loss, train_accuracy

def evaluation(model, dataloader, val_dataset, criterion, device, epoch, num_epochs):
    model.eval()  # лӘЁлҚёмқ„ нҸүк°Җ лӘЁл“ңлЎң м„Өм •
    valid_loss = 0.0
    valid_accuracy = 0

    with torch.no_grad(): # modelмқҳ м—…лҚ°мқҙнҠё л§үкё°
        tbar = tqdm(dataloader)
        for texts, labels in tbar:
            texts = texts.to(device)
            labels = labels.to(device)

            # мҲңм „нҢҢ
            outputs = model(texts)
            loss = criterion(outputs, labels)

            # мҶҗмӢӨкіј м •нҷ•лҸ„ кі„мӮ°
            valid_loss += loss.item()
            # torch.maxм—җм„ң dim мқёмһҗм—җ к°’мқ„ м¶”к°Җн•  кІҪмҡ°, н•ҙлӢ№ dimensionм—җм„ң мөңлҢ“к°’кіј мөңлҢ“к°’м—җ н•ҙлӢ№н•ҳлҠ” мқёлҚұмҠӨлҘј л°ҳнҷҳ
            _, predicted = torch.max(outputs, 1)
            # _, true_labels = torch.max(labels, dim=1)
            valid_accuracy += (predicted == labels).sum().item()


            # tqdmмқҳ м§„н–үл°”м—җ н‘ңмӢңлҗ  м„ӨлӘ… н…ҚмҠӨнҠёлҘј м„Өм •
            tbar.set_description(f"Epoch [{epoch+1}/{num_epochs}], Valid Loss: {loss.item():.4f}")

    valid_loss = valid_loss / len(dataloader)
    valid_accuracy = valid_accuracy / len(val_dataset)

    return model, valid_loss, valid_accuracy


def training_loop(model, train_dataloader, valid_dataloader, train_dataset, val_dataset, criterion, optimizer, device, num_epochs, patience, model_name):
    best_valid_loss = float('inf')  # к°ҖмһҘ мўӢмқҖ validation lossлҘј м ҖмһҘ
    early_stop_counter = 0  # м№ҙмҡҙн„°
    valid_max_accuracy = -1

    for epoch in range(num_epochs):
        model, train_loss, train_accuracy = training(model, train_dataloader, train_dataset, criterion, optimizer, device, epoch, num_epochs)
        model, valid_loss, valid_accuracy = evaluation(model, valid_dataloader, val_dataset, criterion, device, epoch, num_epochs)

        if valid_accuracy > valid_max_accuracy:
            valid_max_accuracy = valid_accuracy

        # validation lossк°Җ к°җмҶҢн•ҳл©ҙ лӘЁлҚё м ҖмһҘ л°Ҹ м№ҙмҡҙн„° лҰ¬м…Ӣ
        if valid_loss < best_valid_loss:
            best_valid_loss = valid_loss
            torch.save(model.state_dict(), f"./model_{model_name}.pt")
            early_stop_counter = 0

        # validation lossк°Җ мҰқк°Җн•ҳкұ°лӮҳ к°ҷмңјл©ҙ м№ҙмҡҙн„° мҰқк°Җ
        else:
            early_stop_counter += 1

        print(f"Epoch [{epoch + 1}/{num_epochs}], Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.4f} Valid Loss: {valid_loss:.4f}, Valid Accuracy: {valid_accuracy:.4f}")

        # мЎ°кё° мў…лЈҢ м№ҙмҡҙн„°к°Җ м„Өм •н•ң patienceлҘј мҙҲкіјн•ҳл©ҙ н•ҷмҠө мў…лЈҢ
        if early_stop_counter >= patience:
            print("Early stopping")
            break

    return model, valid_max_accuracy
````
